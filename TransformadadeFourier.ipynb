```python
import numpy as np
import matplotlib.pyplot as plt
```

Considere o pulso retangular $x(t)$

$$
x(t)= \begin{cases}
			5, & \text{se $-\frac{1}{2}\leq t\leq \frac{1}{2}$}\\
            0, & \text{de outro modo.}
		 \end{cases}
$$  
O código a seguir plota o sinal.

```python
n = 100
t0 = -2
tf = 2
t = np.linspace(t0, tf, n)


x = np.where(np.logical_and(t>=-0.5, t<=0.5), 5, 0) 

# Plota a função   
fg, ax = plt.subplots(1, 1)

ax.plot(t,x, linewidth=2)
ax.grid(True)
plt.ylabel('$x$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

A transformada de Fourier do sinal dada por $X(\omega)$ pode ser computada como
$$
X(\omega)=\int_{-\frac{1}{2}}^{\frac{1}{2}}5e^{-j\omega t}dt,
$$
ou ainda 
$$
X(\omega)=\left.\frac{5}{-j\omega}e^{-j\omega t} \right|_{-\frac{1}{2}}^{\frac{1}{2}}.
$$
Substituindo os limites de integração
$$
X(\omega)=\frac{5}{-j\omega}e^{-j\omega \frac{1}{2}} - \frac{5}{-j\omega}e^{j\omega \frac{1}{2}} =\frac{5}{\omega}\left( \frac{e^{\frac{j\omega}{2}}-e^{\frac{-j\omega}{2}}}{j}\right).
$$
Que pode ser reescrito como
$$
X(\omega)=\frac{5 \times 2}{\omega}\left( \frac{e^{\frac{j\omega}{2}}-e^{\frac{-j\omega}{2}}}{2j}\right)= \frac{5}{\frac{\omega}{2}}\left( \frac{e^{\frac{j\omega}{2}}-e^{\frac{-j\omega}{2}}}{2j}\right)=5 \frac{\sin \left(\frac{\omega}{2} \right)}{\frac{\omega}{2}}=5 \text{sinc}\left(\frac{\omega}{2}\right)
$$


```python
n = 1000
w0 = -2*np.pi
wf = 2*np.pi
w = np.linspace(w0, wf, n)

#A função sinc já é normalizada em função de pi, os pontos em que ela corta o eixo horizontal é múltiplo de pi
X = 5*np.sinc(w/2)

# Plota a função   
fg, ax = plt.subplots(1, 1)

ax.plot(w,np.abs(X), linewidth=2)
ax.grid(True)
plt.ylabel('$|X|$', fontsize=12, rotation=90)
plt.xlabel('$\omega$', fontsize=12)
fg.canvas.draw()
```
