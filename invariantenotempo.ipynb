```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

Verifique se o sistema a seguir é invariante no tempo 
$$
y(t)=3e^{3u(t)}.
$$
Para realizar a verificação de forma gráfica, vamos definir a entrada 
$$u(t)=\sin(t).$$
Na figura a seguir apresenta-se o sinal $y(t)$ bem como a entrada $u(t)$.

```python
t0 = 0   #tempo inicial
tf = 6  #tempo final

n = 1000   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

#Defina uma função de entrada u(t)
u = np.sin(t)

# define a função y(t)
y = np.dot(3,np.exp(np.dot(3,u))) 

# inicia a figura
fg, ax = plt.subplots(2)


#plotando y e u como funções do tempo

ax[0].plot(t,y,'orange', label="$y(t)$", linewidth=2)
ax[1].plot(t,u,'gray', label="$u(t)$", linewidth=2)

#insere uma grade no gráfico
ax[0].grid(True)
ax[1].grid(True)

#posiciona a legenda na melhor localização
ax[0].legend(loc='best')
ax[1].legend(loc='best')

#insere informação do eixo vertical
ax[0].set(ylabel='$y(t)$')
ax[1].set(ylabel='$u(t)$')

#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Vamos avaliar qual o efeito produzido por um atraso na entrada. Ou seja, estamos avaliando $G(u(t-\tau))$. Para $\tau= 2$ tem-se:

```python
t0 = 0   #tempo inicial
tf = 6  #tempo final
tau = 2 # valor do atraso
n = 1000   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

#Defina uma função de entrada u(t)
ut = np.sin(t-tau)

# define a função y(t)
yt = np.dot(3,np.exp(np.dot(3,ut))) 

# inicia a figura
fg, ax = plt.subplots(2)

#plotando as variáveis atrasadas como funções do tempo

ax[0].plot(t,yt,'orange', label=r"$G[u(t-\tau)]$", linewidth=2)
ax[1].plot(t,ut,'gray', label="$u(t)$", linewidth=2)

#insere uma grade no gráfico
ax[0].grid(True)
ax[1].grid(True)

#posiciona a legenda na melhor localização
ax[0].legend(loc='best')
ax[1].legend(loc='best')

#insere informação do eixo vertical
ax[0].set(ylabel=r"$G[u(t-\tau)]$")
ax[1].set(ylabel=r"$u(t-\tau)$")

#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Observe, que a resposta para a entrada atrasada preserva o mesmo formato da resposta para a entrada original. Além disso, ela está deslocada de exatamente duas unidades de tempo. 


Note que aplicando um atraso na saída tem-se
$$
y(t-\tau)=3e^{3u(t-\tau)}.
$$
Além disso,
$$u(t-\tau)=\sin(t-\tau).$$
Desta forma, concluímos que 
$$
y(t-\tau)=G[u(t-\tau)].
$$
Confirmando que o sistema é invariante no tempo.
