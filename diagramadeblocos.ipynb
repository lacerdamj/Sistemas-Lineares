Encontre a relação entre a entrada $R$ e saída $C$

```python
#Importando as bibliotecas
import sympy
sympy.init_printing()
```

O primeiro passo é definir o conjunto de variáveis do problema. Serão criadas variáveis auxiliares $V_1,\ldots V_9,$ para definir as relações entre os blocos, e também relações entre os somadores. A Figura a seguir apresenta o conjunto de variáveis auxiliares e sua posição no diagrama.

Diagramadeblocosvaraux.jpg

Na sequência todas as variáveis são declaradas como variáveis simbólicas. Note que as variáveis desconhecidas são os sinais auxiliares $V$ e a saída. As funções dos blocos e a entrada $R$ são variáveis conhecidas. 

```python
(R, V1, V2, V3, V4, V5, V6, V7, V8, V9, C,
 G1, G2, G3, G4, H1, H2, H3) = sympy.symbols('R, V1, V2, V3, V4, V5, V6, V7, V8, V9, C,'
                                         'G1, G2, G3, G4, H1, H2, H3')
unknowns = V1, V2, V3, V4, V5, V6, V7, V8, V9, C
```

As equações que definem as relações entre os blocos e entre os somadores são declaradas.

```python
eqs = [# Blocks
       V2 - G1*V1,
       V4 - G2*V3,
       V6 - G3*V5,
       C - G4*V6,
       V7 - H2*C,
       V8 - H3*V6,
       V9 - H1*V4,
       # Sums
       V1 - (R - V9),
       V3 - (V2 - V8),
       V5 - (V4 - V7),
       ]
```

O comando sympy.solve resolve às equações considerando as variáveis desconhecidas que foram declaradas. A solução relaciona cada uma das variáveis deconhecidas com as variáveis conhecidas $R$ e as funções $G$ e $H$.

```python
sol = sympy.solve(eqs, unknowns)
sol
```

Nosso interesse é somente na relação entre a entrada $R$ e a saída $C$, logo:

```python
sol[C].factor()
```

Para obter a função de transferência da entrada para a saída, basta dividir a expressão por $R$. Desta forma tem-se

```python
print("A função de transferência é") 

sol[C].factor()/R
```
