```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

Considere o seguinte sistema
$$
y(t)=2\frac{du(t)}{dt}.
$$
Verifique se o sistema é linear. 

Para verificar se o sistema é linear vamos calcular 
$$
y_1(t)=G[u_1(t)], \quad y_2(t)=G[u_2(t)].
$$
em que $G[\cdot{}]$ é o operador do sistema, nesse caso o operador derivada.

Para ilustrar de forma gráfica o princípio da superposição, dois sinais foram escolhidos:

$$u_1(t)= \sin(t), \quad u_2(t)= \exp(-2t). $$ 

Os sinais escolhidos são ilustrados na próxima Figura.

```python
import numpy as np
from matplotlib import pyplot as plt

#tempo inicial
t0 = 0
#tempo final
tf = 2*np.pi
# passo considerado
dt = np.pi/100

t = np.arange(t0,tf,dt)

# define as funções u1(t) e u2(t)
u1 = np.sin(t) 
u2 = np.exp(-2*t)

fg, ax = plt.subplots(1, 1)

#Plota as funções
ax.plot(t,u1,'green', label="$u_1(t)$", linewidth=2)
ax.plot(t,u2,'red', label="$u_2(t)$", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Neste ponto vamos verificar o efeito de cada sinal da entrada na saída do sistema.
$$
y_1(t)=G[u_1(t)]=2\frac{du_1(t)}{dt}, \quad y_2(t)=G[u_2(t)]=2\frac{du_2(t)}{dt}.
$$

```python
# Para calcular as saídas y_1 e y_2, vamos empregar a função 
# gradient do numpy, que deriva o sinal de acordo com o passo dt considerado.

y1 = np.gradient(2*np.sin(t), dt)
y2 = np.gradient(2*np.exp(-2*t), dt)

fg, ax = plt.subplots(1, 1)

#Plota as funções
ax.plot(t,y1,'green', label="$y_1(t)$", linewidth=2)
ax.plot(t,y2,'red', label="$y_2(t)$", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Para atender o princípio da superposição, o sinal deve satisfazer as propriedades da homogeneidade e da aditividade. Para testar as duas propriedades de uma só vez, consideramos uma combinação linear das saídas:

$$y(t)=\alpha_1y_1(t) +\alpha_2y_2(t),$$ 
com $\alpha_1=2$ e $\alpha_2=3$. Desta forma temos o seguinte resultado.

```python
#Define a combinação das saídas
yc=2*y1+3*y2

fg, ax = plt.subplots(1, 1)

#Plota o sinal
ax.plot(t,yc,'black', label=r"$\alpha_{1}y_1(t)+\alpha_{2}y_2(t)$", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Para verificar se o sistema é linear, devemos ponderar as entradas considerando os mesmos pesos, e verificar o efeito das entradas combinadas na saída do sistema.
$$
G\left[\alpha_1u_1(t)+ \alpha_2u_2(t)\right.].
$$
Os mesmos pesos devem ser considerados, ou seja $\alpha_1=2$ e $\alpha_2=3$.
O código a seguir computa o efeito na saída do sistema.


```python
# Sinal de entrada ponderado
up=2*np.sin(t)+3*np.exp(-2*t)

#aplica o operador do sistema, neste caso a derivada.
yp = np.gradient(2*up, dt)

fg, ax = plt.subplots(1, 1)

#Plota o sinal resultante
ax.plot(t,yp,'m', label=r"$G[\alpha_{1}u_{1}(t)+ \alpha_{2}u_{2}(t)]$", linewidth=2)


ax.grid(True)
ax.legend(loc='best')
plt.xlabel(r"$t$", fontsize=12)
plt.ylabel(r"$G[\alpha_{1}u_{1}(t)+ \alpha_{2}u_{2}(t)]$")
fg.canvas.draw()
```

Por fim, vamos plotar as duas curvas no mesmo gráfico, para verificar que a seguinte igualdade é verificada. 
$$
\alpha_1y_1(t)+ \alpha_2y_2(t)= G[\alpha_1u_1(t)+ \alpha_2u_2(t)].
$$



```python
fg, ax = plt.subplots(1, 1)

ax.plot(t,yp,'r', label=r"$G[\alpha_{1}u_{1}(t)+ \alpha_{2}u_{2}(t)]$", linewidth=2)
ax.plot(t,yc,'y--', label=r"$\alpha_{1}y_1(t)+\alpha_{2}y_2(t)$", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Logo, o sistema 
$
\displaystyle{y(t)=2\frac{du(t)}{dt}},
$
é um sistema linear.
