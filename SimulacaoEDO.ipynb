```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```


Considere o sistema  

$$\displaystyle{\frac{d y(t)}{dt}+2y(t)}=4u(t).$$
Considerando condições iniciais nulas:

1.   Plote a resposta do sistema para $u_1(t)=1$.
2.   Plote a resposta do sistema para $u_2(t)=\sin(3t)$.
3.   Plote a resposta para a entrada $u(t)=u_1(t)+u_2(t)$ e verifique se o sistema é linear.

No primeiro passo vamos encontrar a resposta para a entrada $u_1(t)$. Ou seja, vamos determinar 

$$
y_1(t)=G[u_1(t)],
$$

em que $G$ representa o sistema.

```python
# Para solucionar a EDO precisamos definir o modelo do sistema
def model(y,t):
    u1 = 1
    dydt = -2 * y+4*u1
    return dydt

# condição inicial nula
y0 = 0

t0 = 0     #tempo inicial
tf = 3     #tempo final
n = 1000   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# Resolve a EDO
y1 = odeint(model,y0,t)

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável y
ax.plot(t,y1,'r', label="$y_1(t)$", linewidth=2)

#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$y_1(t)$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

No segundo passo vamos encontrar a resposta para a entrada $u_2(t)$. Ou seja, vamos determinar 

$$
y_2(t)=G[u_2(t)],
$$

em que $G$ representa o sistema.

```python
# Para solucionar a EDO precisamos definir o modelo do sistema
def model(y,t):
    u2 = np.sin(3*t)
    dydt = -2 * y+4*u2
    return dydt

# condição inicial nula
y0 = 0

t0 = 0     #tempo inicial
tf = 3     #tempo final
n = 1000   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# Resolve a EDO
y2 = odeint(model,y0,t)

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável y
ax.plot(t,y2,'k', label="$y_2(t)$", linewidth=2)

#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$y_2(t)$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

No terceiro passo vamos encontrar a resposta para a entrada $u(t)=u_1(t)+u_2(t)$:

$$
y(t)=G[u(t)].
$$

Vamos verificar se o sistema respeita o princípio da aditividade. Devemos confirmar a seguinte equivalência:
$$
G[u_1(t)+u_2(t)]=y_1(t)+y_2(t)
$$

```python
# Para solucionar a EDO precisamos definir o modelo do sistema
def model(y,t):
    u = np.sin(3*t)+1
    dydt = -2 * y+4*u
    return dydt

# condição inicial nula
y0 = 0

t0 = 0     #tempo inicial
tf = 3     #tempo final
n = 1000   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# Resolve a EDO
y = odeint(model,y0,t)

# inicia a figura
fg, ax = plt.subplots(1, 1)


#plota a soma das saídas y_1 e y_2 para vericar se o sistema é linear
ax.plot(t,y1+y2,'y', label="$y_1(t)+y_2(t)$", linewidth=2)

#plota no eixo horizontal a variável t e no vertical a variável y
ax.plot(t,y,'b--',alpha=0.5, label=r"$G[u_1(t)+u_2(t)]$", linewidth=2)


#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')

#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Na sequência vamos verificar se o sistema respeita também o princípio da homogeneidade. Em outras palavras, vamos verificar a igualdade

$$
G[\alpha_1 u_1(t)+\alpha_2 u_2(t)]=\alpha_1 y_1(t)+\alpha_2y_2(t).
$$

Para a simulação foram escolhidos $\alpha_1=2$ e $\alpha_2=3$, sem perda de generalidade.

```python
# Para solucionar a EDO precisamos definir o modelo do sistema
def model(y,t):
  alfa1 = 2
  alfa2 = 3
  u = alfa1*1+alfa2*np.sin(3*t)
  dydt = -2 * y+4*u
  return dydt

# condição inicial nula
y0 = 0
t0 = 0     #tempo inicial
tf = 3     #tempo final
n = 1000   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# Resolve a EDO
y = odeint(model,y0,t)

# inicia a figura
fg, ax = plt.subplots(1, 1)

alfa1 = 2
alfa2 = 3

#plota a soma das saídas y_1 e y_2 para vericar se o sistema é linear
ax.plot(t,alfa1*y1+alfa2*y2,'y', label=r"$\alpha_1y_1(t)+\alpha_2y_2(t)$", linewidth=2)

#plota no eixo horizontal a variável t e no vertical a variável y
ax.plot(t,y,'b--',alpha=0.5, label=r"$G[\alpha_1 u_1(t)+\alpha_2 u_2(t)]$", linewidth=2)


#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')

#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Note que, o último passo seria suficiente para verificar se o sistema é linear, uma vez que ele contempla o princípio da aditividade e também o princípio da homogeneidade.
