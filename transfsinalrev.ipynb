```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

### Reversão no tempo. 

Considere o seguinte sinal
$$g(t)= \left\{ \begin{aligned} 
e^{-2t} & \quad 0 \leq t \leq 2 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. $$ 

O sinal $g(t)$ pode ser plotado da seguinte forma:

```python
n = 10000   #número de pontos
t0 = -3        #tempo inicial
tf = 3      #tempo final
#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# define a função g(t)
g = np.where(np.logical_and(t>=0, t<=2), np.exp(-2*t), 0) 

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável x
# 'b' de blue define a cor azul.
ax.plot(t,g,'b', label="$g(t)$", linewidth=2)
#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$g(t)$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

O sinal $g(-t)$ é a imagem espelhada da função $g(t)$ com relação ao eixo vertical. O sinal $g(-t)$ pode ser obtido da seguinte forma:

$$g(-t)= \left\{ \begin{aligned} 
e^{-2(-t)} & \quad 0 \leq -t \leq 2 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. = 
\left\{ \begin{aligned} 
e^{2t} & \quad -2 \leq t \leq 0 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right.$$ 

O sinal $g(-t)$ é mostrado na próxima Figura.

```python
#define a função g(-t)
gm = np.where(np.logical_and(-t>=0, -t<=2), np.exp(-2*(-t)), 0) 
   
fg, ax = plt.subplots(1, 1)
ax.plot(t,gm,'k--', label="$g(-t)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$g(-t)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

O efeito do espelhamento em relação ao eixo vertical fica claro quando plotamos as duas figuras no mesmo gráfico.

```python
fg, ax = plt.subplots(1, 1)

ax.plot(t,g,'b', label="$g(t)$", linewidth=2)
ax.plot(t,gm,'k--', label="$g(-t)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```
