```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

### Deslocamento no tempo

Considere o seguinte sinal

$$f(t)= \left\{ \begin{aligned} 
t^2-1 & \quad 1 \leq t \leq 3 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. $$ 

```python
n = 10000   #número de pontos
t0 = 0        #tempo inicial
tf = 6      #tempo final
#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# define a função f(t)
f = np.where(np.logical_and(t>=1, t<=3), t**2-1, 0) # build input functions

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável x
# 'b' de blue define a cor azul.
ax.plot(t,f,'b', label="$f(t)$", linewidth=2)
#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$f(t)$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Para plotar o sinal $f(t-1)$, deve-se alterar $t$ por $t-1$ na função original. Neste caso, tem-se

$$f(t-1)= \left\{ \begin{aligned} 
(t-1)^2-1 & \quad 1 \leq t-1 \leq 3 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. = \left\{ \begin{aligned} 
t^2-2t & \quad 2 \leq t \leq 4 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right.$$

Note que, quando deslocamos a função original, o intervalo no qual ela é avaliada também muda. Como resultado, a função original vai deslocar para a direita neste caso, como mostrado na próxima Figura.

```python
#define a função f(t-1)
f1 = np.where(np.logical_and(t-1>=1, t-1<=3), (t-1)**2-1, 0) # build input functions
   
fg, ax = plt.subplots(1, 1)
ax.plot(t,f1,'orange', label="$f(t-1)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$f(t-1)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Para plotar o sinal $f(t+1)$, devemos alterar $t$ por $t+1$ na função original. Neste caso obtemos:

$$f(t+1)= \left\{ \begin{aligned} 
(t+1)^2-1 & \quad 1 \leq t+1 \leq 3 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. = \left\{ \begin{aligned} 
t^2+2t & \quad 0 \leq t \leq 2 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right.$$ 

Neste cenário, a função original se desloca para a esquerda, como apresentado na Figura a seguir.

```python
#define a função f(t+1)
f2 = np.where(np.logical_and(t+1>=1, t+1<=3), (t+1)**2-1, 0) # build input functions


fg, ax = plt.subplots(1, 1)
ax.plot(t,f2,'brown', label="$f(t+1)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$f(t+1)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Por fim, vamos plotar todos os gráficos em uma única figura para visualizar o efeito do deslocamento no tempo para o sinal $f(t)$.

```python
fg, ax = plt.subplots(1, 1)

ax.plot(t,f, 'blue', label="$f(t)$", linewidth=2)
ax.plot(t,f1, 'orange', label="$f(t-1)$", linewidth=2)
ax.plot(t,f2, 'brown', label="$f(t+1)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```
