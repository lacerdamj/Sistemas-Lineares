```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

### Sinais contínuos no tempo

Considere o seguinte sinal
$$g(t)= \sin(t) $$ 

Para representar um sinal contínuo no tempo, devemos escolher um número de pontos que permita a correta caracterização do sinal.

```python

t0 = 0   #tempo inicial
tf = 2*np.pi  #tempo final

n1 = 5   #número de pontos
n2 = 50   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t1 = np.linspace(t0, tf, n1)
t2 = np.linspace(t0, tf, n2)

# define a função g(t)
g1 = np.sin(t1) 
g2 = np.sin(t2)

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável x
# 'b' de blue define a cor azul.
ax.plot(t1,g1,'b', label="$g1(t)$", linewidth=2)
ax.plot(t2,g2,'r', label="$g2(t)$", linewidth=2)

#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$g(t)$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

###Sinais discretos no tempo

Para exemplificar vamos apresentar a versão discreta do sinal anterior. Neste caso, as amostras serão realizadas em intervalos de tempo igualmente espaçados.


```python
t0 = 0   #tempo inicial
tf = 6   #tempo final

n = 7   #número de pontos

#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# define a função g[n]
g = np.sin(t) 

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável x
# 'b' de blue define a cor azul.
plt.stem(t,g,'m', label="$g[n]$")
#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$g[n]$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('n', fontsize=12)
fg.canvas.draw()
```

```python
fg, ax = plt.subplots(1, 1)

plt.xlim([0, 7])
ax.plot(t2,g2,'r', label="$g(t)$", linewidth=2)
plt.stem(t,g,'m', label="$g[n]$")

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('n/t', fontsize=12)
fg.canvas.draw()
```
