```python
import numpy as np
import matplotlib.pyplot as plt
```

Considere um sinal com período $T_0=2s$. A expressão do sinal em um ciclo é dada por:

$$
x(t)= \begin{cases}
			1-t, & \text{se $0<t\leq 1s$}\\
            0, & \text{se $1 \leq t\leq 2s$}
		 \end{cases}
$$  
A Figura a seguir plota o sinal ao longo de dois ciclos.

```python
n = 100
tf = 4
t = np.linspace(0, tf, n)


x1 = np.where(np.logical_and(t>=0, t<=1), 1-t, 0) # construindo o primeiro ciclo
x2 = np.where(np.logical_and(t>=2, t<=3), 3-t, 0) # construindo o segundo ciclo

# O sinal total é a soma dos ciclos
x=x1+x2 

# Plota a função   
fg, ax = plt.subplots(1, 1)

ax.plot(t,x, linewidth=2)
ax.grid(True)
plt.ylabel('$x$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

A representação amplitude/fase da série de Fourier é dada por

$$
f(t)=c_0 +\sum_{n=1}^{\infty} c_n\cos(n\omega_0t+\phi_n),
$$
em que $c_n$ é a amplitude do harmônico $n$, e $\phi_n$ é a fase associada. Neste exemplo vamos plotar a série para os $4$ primeiros harmônicos. A ideia é avaliar a contribuição da inclusão de cada harmônico.


```python
c0=0.25; 
c1=0.38; phi1=-57.5
c2=0.16; phi2=-90
c3=0.11; phi3=-78
c4=0.08; phi4=-90
w0=np.pi

xf1=c0+c1*np.cos(1*w0*t+np.deg2rad(phi1))
xf2=xf1+ c2*np.cos(2*w0*t+np.deg2rad(phi2))
xf3=xf2+ c3*np.cos(3*w0*t+np.deg2rad(phi3))
xf4=xf3+c4*np.cos(4*w0*t+np.deg2rad(phi4))


fg, ax = plt.subplots(1, 1)

ax.plot(t,xf1, label="$x_{f1}$", linewidth=2)
ax.plot(t,xf2,'y.', label="$x_{f2}$", linewidth=2)
ax.plot(t,xf3,'g--', label="$x_{f3}$", linewidth=2)
ax.plot(t,xf4,'r',label="$x_{f4}$", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Vamos plotar as duas curvas no mesmo gráfico, para verificar se a série de Fourier foi capaz de representar bem a função original.

```python
fg, ax = plt.subplots(1, 1)

ax.plot(t,x,'r', label=r"Série de Fourier", linewidth=2)
ax.plot(t,xf4,'b--', label=r"Sinal original", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

```python
#Espectro de amplitude

c=[c0,c1,c2,c3,c4]
tc = np.linspace(0, 4,5)

fg, ax = plt.subplots(1, 1)
ax.stem(tc, c)
plt.xlabel('Espectro de amplitude', fontsize=12)
plt.ylabel('$c_n$', fontsize=12)
ax.grid(True)

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = '0'
labels[2] =''
labels[3] ='$\omega_0$'
labels[4] =''
labels[5] ='$2\omega_0$'
labels[6] =''
labels[7] ='$3\omega_0$'
labels[8] =''
labels[9] ='$4\omega_0$'

ax.set_xticklabels(labels)
fg.canvas.draw()
```
